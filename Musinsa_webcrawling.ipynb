```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
from tqdm import tqdm
from selenium.webdriver.common.keys import Keys
import urllib.request
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options = options)
driver.implicitly_wait(3)
driver.get("https://www.musinsa.com/app/styles/lists") #무신사 링크
time.sleep(3)
#driver.find_element(By.XPATH,'').click()

image_list = []
image_name_list = []

address = r"C:\무신사 크롤링 이미지\golf"
```

```python
# 스타일별
from selenium.webdriver.common.by import By
for style in range(14,15): # (range(1,10))
    driver.find_element(By.XPATH,f'//*[@id="catelist"]/div[3]/div/dl/dd/ul/li[{style}]/a/span').click()
    
    #페이지별
    for page in tqdm(range(1,12)): #(range1 ~ 10 page)
        if (page != 1):
            driver.find_element(By.XPATH,f'/html/body/div[3]/div[2]/form/div[4]/div/div[4]/div/div/a[{page+1}]').send_keys(Keys.ENTER)
                
        # 이미지 리스트별
        for image_idx in tqdm(range(1,61)): # range(1,61)
                        
            # 이미지 클릭해서 하나씩 가져오기
            try : 
                # 오류나면 화면 크게 해줘야함!
                driver.find_element(By.CSS_SELECTOR,f"body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({image_idx}) > div.style-list-item__thumbnail > a > div > img").click()
                time.sleep(0.5)
                
                imgurl = driver.find_element(By.CSS_SELECTOR,'#style_info > div.detail_slider_wrap > div.detail_slider.slick-initialized.slick-slider > div > div > img.detail_img.slick-slide.slick-current.slick-active').get_attribute("src")
                image_list.append(imgurl)

                image_name = imgurl.split('/')[-1]
                image_name_list.append(image_name)            

                urllib.request.urlretrieve(imgurl,address+"/"+str(image_name)) # download images in address folder
                
                # 이전페이지로 돌아가기
                driver.back()
                time.sleep(1)
            
            # 비디오 예외
            except NoSuchElementException:
                
                driver.back()
```
